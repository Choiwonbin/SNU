## HW2. Line & Stacked Bar Plot

```python
# 필요한 library import
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plot할 excel을 dataframe으로 불러오기
df = pd.read_excel('last_amb_hour_day.xls', skiprows=5, header=None,
                   names = ['date', 'PM10', 'PM2.5', 'O3', 'NO2', 'CO', 'SO2', 'Confirm']).sort_values(['date'], ascending=True).reset_index()
del df['index']
```

```python
# date 중 일자만 뽑아서 day column 생성
df['day'] = df['date'].astype(str).str[8:]
```

```python
# plot
plt.figure(figsize = (8, 16))

plt.subplot(211)
plt.plot(df['day'], df['PM10'])
plt.plot(df['day'], df['PM2.5'])
plt.xlabel('Date')
plt.ylabel('Concentration($ug/cm^3$)')
plt.title('Fine dust change in February 2018')
plt.legend(loc=2)

plt.subplot(212)
plt.bar(df['day'], df['O3'])
plt.bar(df['day'], df['NO2'], bottom = df['O3'])
plt.bar(df['day'], df['SO2'], bottom = df['O3'] + df['NO2'])
plt.xlabel('Date')
plt.ylabel('Concentration(ppm)')
plt.title('$O_3, NO_2, SO_2$ change in February 2018')
plt.legend(['$O_3$', '$NO_2$', '$SO_2$'], loc=2)

plt.show()
```
