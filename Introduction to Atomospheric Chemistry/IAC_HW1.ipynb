## HW1. Contour Plot

- Left
\begin{equation*}
Z = sinX + sinY
\quad(X = [-2\pi, 2\pi], Y = [-2\pi, 2\pi])
\end{equation*}

- Right
\begin{equation*}
Z = Xe^{-X^2-Y^2}
\quad(X = [-2, 2], Y = [-2, 2])
\end{equation*}

```python
# 필요한 library import
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot할 수식, 축범위 생성
a = np.linspace(-2*np.pi,2*np.pi)
b = np.linspace(-2*np.pi,2*np.pi)
A,B = np.meshgrid(a,b)

d = np.linspace(-2,2)
e = np.linspace(-2,2)
D,E = np.meshgrid(d,e)

X = np.sin(A) + np.sin(B)
Y = D*np.exp(-D**2-E**2)
```

```python
# Contour plot
plt.figure(figsize = (8,4))

plt.subplot(121)
C = plt.contour(A,B,X,10, cmap=plt.cm.Reds, linewidth=0.5)
plt.clabel(C, inline=1, fontsize=10)
plt.title('Z = sinX + sinY')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(122)
C = plt.contour(D,E,Y,10, cmap=plt.cm.Blues, linewidth=0.5)
plt.clabel(C, inline=1, fontsize=10)
plt.title(r'$Z=Xe^{-X^2-Y^2}$')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
